```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

```

```python
df = pd.read_csv('netflix_titles.csv')
df.head(10)
```

Handling missing values

```python
df['description'] = df['description'].fillna('')
df['director'] = df['director'].fillna('')
df['cast'] = df['cast'].fillna('')
df['listed_in'] = df['listed_in'].fillna('')
df.head(10)
```

Directors with most movies

```python
director_counts = df['director'].replace('', None).dropna().str.split(', ').explode().value_counts().head(10)
print(director_counts)
# Plot the bar chart
plt.figure(figsize=(10, 5))
colors = plt.get_cmap('tab10', len(director_counts))
director_counts.plot(kind='bar', color=colors(range(len(director_counts))))

# Formatting
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.title("Top 10 Directors with Most Movies on Netflix")
plt.xticks(rotation=45)
plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()  
movies['duration_num'] = movies['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(12, 6))
hist = sns.histplot(data=movies, x='duration_num', bins=50, kde=False, color="red", edgecolor="black")
heights = hist.patches

# Apply gradient color based on the height of each bar
for patch in heights:
    # Normalize the height for color scaling
    height = patch.get_height()
    color = plt.cm.Greens(1 - height / max([patch.get_height() for patch in heights]))  
    patch.set_facecolor(color)
plt.title('Movie Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')

plt.show()
```
